# Euro Exchange Project

The dataset we'll use describes Euro daily exchange rates between 1999 and 2022. The euro (symbolized with â‚¬) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Our focus in the guided part of the project will be on the exchange rate between the euro and the American dollar.

### Data Cleaning
We will rename the '[US dollar]' columns and 'Period\Unit:' columns to make it easier to work with.

```python
df.columns
```

```python
df.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'}, #pay attention here tow inverted dashes
                      inplace=True)

df['Time'] = pd.to_datetime(df['Time'])
df.sort_values('Time', inplace=True) #sorting values
df.reset_index(drop=True, inplace=True)
df.head()
```

Isolating the 'Time' and the 'US_dollar' columns:

```python
euro_to_dollar = df[['Time', 'US_dollar']].copy()
df['US_dollar'].value_counts()
```

```python
#dropping the values that start with '-' and transforming it to float
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

### Roling Mean





```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

### Coming up with an Idea

How the euro-dollar rate has changed during the coronavirus pandemic  compared to previous years.

```python
#separating the years
euro_to_dollar['year'] = euro_to_dollar['Time'].dt.year

previous_year = euro_to_dollar.copy()[(euro_to_dollar['year'] >= 2016) & (euro_to_dollar['year'] < 2020)]
print(previous_year.shape)

pandemic_year = euro_to_dollar.copy()[(euro_to_dollar['year'] >= 2020)]
print(pandemic_year.shape)
```

## Euro_Dollar exchange rate during **COVID-19**:

```python
#FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

#Adding the plot
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(previous_year['Time'],
        previous_year['rolling_mean'],
        linewidth=1, color='#A6D785'
        )

#Highlighting the year of the pandemy
ax.plot(pandemic_year['Time'],
        pandemic_year['rolling_mean'],
        linewidth=3, color='#e23d28',
        )

#Highlight the peak of pandemy
ax.axvspan(xmin=18580, xmax=18870, ymin=.09,
           alpha=.3, color='grey')


ax.text(16500.0, 1.3, 'Euro-USD rate peaked at 1.22 during 2019-2022Â´s Covid-19 pandemic',
        weight='bold', size=12)
ax.text(16500.0, 1.27, 'Euro-USD exchange rates between 2016 and 2023',
      size=12)

ax.tick_params(axis='x', colors='grey')
ax.tick_params(axis='y', colors='grey')

#Adding signature
ax.text(16500, .9, 'Henrique Araujo' +' '*55 + 'Source: European Central Bank',
        color='#f0f0f0', backgroundcolor='#4d4d4d')
plt.show()
```

Apparently the covid itself did not interfer that much in the Euro-Dollar exchange rate. I thought the COVID-19 situation would change drastically the Euro-USD exchange rates, but aparrently it doesnÂ´t.

## Financial Crisis Graphic:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Separating the crisis from normal date
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006
                                         ) & (euro_to_dollar['Time'].dt.year <= 2009)]

financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007
                                              ) & (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
#Adding the plot
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='#A6D785'
        )

#Highlighting the year of the pandemy
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#e23d28',
        )
### Highlihting the peak of the crisis
ax.axvspan(xmin=13950, xmax=14150, ymin=.09,
           alpha=.3, color='grey')

#Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x=13100
for year in ['2006', '2007', '2008', '2009', '2010']:
  ax.text(x, 1.13, year, alpha=.5, fontsize=11)
  x += 365

y = 1.197
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(13000, y, rate, alpha=0.5, fontsize=11)
    y += 0.1

#Adding title and subtitle
ax.text(13000, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
   weight='bold')

ax.text(13000, 1.63, 'Euro-USD exchange rates between 2006 and 2010', size = 12)

#Adding signature
ax.text(13000, 1.08, 'Henrique Araujo' + ' '*94 + 'Source: European Central Bank',
        color='#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

#Add some transparency to the grid
ax.grid(alpha=.5)

plt.show()
```

By analyzing the graph, we can notice that after the 2007-2008Â´s economic crisis in USA, that was a significant increase in the Euro-USD exchange rate, due to the worsening of AmericaÂ´s economy.

## The Three US Presidencies Graphic:

```python
#Separating the presidents by their mandate
three_presid = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2022)]

bush = three_presid.copy()[(three_presid['Time'].dt.year < 2009)]

obama = three_presid.copy()[(three_presid['Time'].dt.year >= 2009) & (three_presid['Time'].dt.year < 2017)]

trump = three_presid.copy()[(three_presid['Time'].dt.year >= 2017) & (three_presid['Time'].dt.year < 2021)]
```

```python
style.use('fivethirtyeight')

#Adding subplots
plt.figure(figsize=(12,6))
ax1= plt.subplot(2,3,1)
ax2= plt.subplot(2,3,2)
ax3= plt.subplot(2,3,3)
ax4= plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

#Changes to all the subplots
#Creating a pattern for all the plots
for ax in axes:
  ax.set_ylim(.8, 1.7) #setting the size of the y
  ax.set_yticks([1.0, 1.2, 1.4, 1.6])
  ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'],
                     alpha=.3)
  ax.grid(alpha=.5)

#Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'],
         color='#BF5FFF')
ax1.set_xticklabels(['', '2001','', '2003','', '2005', '',
                     '2007', '', '2009'],
                    alpha=.3)
#Adding title and subtitle
ax1.text(12400, 1.85, 'BUSH', fontsize=18, weight='bold',
         color='#BF5FFF')
ax1.text(12200, 1.75, '2001-2009', weight='bold',
         alpha=.3)


#Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax2.set_xticklabels(['','2009','','2011','','2013','',
                    '2015','','2017'],
                    alpha=.3)
#Adding title and subtitle
ax2.text(15200, 1.85, 'OBAMA', fontsize=18, weight='bold',
         color='#ffa500')
ax2.text(15150, 1.75, '2009-2017', weight='bold',
         alpha=.3)

#Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'],
         color='#00B2EE')
ax3.set_xticklabels(['2017','','2018','','2019','',
                     '2020','','2021'],
                    alpha=.3)
#Adding title and subtitle
ax3.text(17660, 1.85, 'TRUMP', fontsize=18, weight='bold',
         color='#00B2EE')
ax3.text(17600, 1.75, '2017-2021', weight='bold',
         alpha=.3)

#Creating graphic of the three presidents
#Ax4: All
ax4.plot(bush['Time'], bush['rolling_mean'],
         color='#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'],
         color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'],
         color='#00B2EE')
ax4.grid(alpha=.5)
ax4.set_xticks([])

#Adding title and subtitle for the entire graphic:
ax1.text(10800, 2.27, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(10800, 2.07, '''EURO-USD rates under Geroge W. Bush (2001 - 2009), Barack Obama(2009 - 2017) and
Donald Trump (2017-2021)''',
         fontsize=16)

#Adding my signature
ax4.text(10650, 0.65, 'Henrique Araujo' + ' '*112 + 'Source: European Central Bank',
         color='#f0f0f0', backgroundcolor ='#4d4d4d',
         size=14)
plt.show()
```
